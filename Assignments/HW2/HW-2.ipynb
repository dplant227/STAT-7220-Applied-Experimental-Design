```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
import openpyxl
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan


#p1 Import Data#
dstm = pd.read_excel("Data Science Teaching Method.xlsx", engine = 'openpyxl')

#p1 Question 6 Exploratory Data Analysis.
print(dstm.columns.tolist())
print(dstm.describe(include="all").T)

print(dstm.info)

#specifying order

know_order = ['Low', 'High']

dstm['Prior_Knowledge'] = pd.Categorical(dstm['Prior_Knowledge'], categories=know_order, ordered =True)

#plot
sns.set_style("whitegrid")

sns.boxplot(
    data=dstm,
    x='Teaching_Method',
    y='Score',
    hue='Prior_Knowledge',
    hue_order=know_order, #order it nice
    palette = 'pastel'
)
plt.title("Effect of Teaching Method on Scores\nby Prior Knowledge",
          loc='center',
          fontsize=16)
plt.show()

#p1 Question 7 ANOVA
#fit model
dstm_mod = ols('Score ~ Teaching_Method + Prior_Knowledge', data=dstm).fit()
print(dstm_mod.summary())

#obtain anova table
dstm_table = sm.stats.anova_lm(dstm_mod, type=1)
print(dstm_table)

#Add significance column
dstm_table["significant"] = np.where(dstm_table["PR(>F)"] < 0.05, "Significant", "Not Significant")
print(dstm_table.loc[["Teaching_Method", "Prior_Knowledge"]])
#Check for Normality
dstm_residuals = dstm_mod.resid

#QQ plot

sm.qqplot(dstm_residuals, line='s')
plt.title("QQ Plot of Residuals")
plt.show()

#Shapiro Wilk test
shapiro_test = stats.shapiro(dstm_residuals)
print(shapiro_test)

#Test stat
shapiro_stat = shapiro_test.statistic

#p-value
shapiro_p = shapiro_test.pvalue

print(f"Shapiro-Wilk Test Statistic: {shapiro_stat:.4f}")
print(f"Shapiro-Wilk Test P-Value: {shapiro_p:.4f}")
if shapiro_p < 0.05:
    print("The residuals are not normally distributed (reject H0).")
else:
    print("The residuals are normally distributed (fail to reject H0).")

#p1 Question 8 Homogeneity of Variance

#Standardize residuals

standardized_residuals = dstm_mod.get_influence().resid_studentized_internal

fitted_values = dstm_mod.fittedvalues

#Create plot

plt.figure(figsize=(8,6))
sns.scatterplot(x=fitted_values, y=standardized_residuals)
plt.axhline(0, color='red', linestyle='--')
plt.axhline(3, color='blue', linestyle='--')
plt.axhline(-3, color='blue', linestyle='--')
plt.title("Standardized Residuals vs. Fitted Values", fontsize=16)
plt.xlabel("Fitted Values")
plt.ylabel("Standardized Residuals")
plt.show()

#B-P test

bp_test = het_breuschpagan(dstm_mod.resid, dstm_mod.model.exog)

#F-stat

bp_stat = bp_test[0]

#p-value

bp_p = bp_test[1]

print(f"F-statistic: {bp_stat}")
print(f"F p-value: {bp_p}")
if bp_p < 0.05:
    print("Heteroscedasticity detected (reject H0). Constant Variance not assumed.")
else:
    print("No heteroscedasticity detected (fail to reject H0). Constant Variance assumed.")


#p1 Question 9
#Print Anova Table

print(dstm_table.loc[["Teaching_Method", "Prior_Knowledge"]])

#p1 Question 10 Tukey HSD

tukey_result = pairwise_tukeyhsd(endog=dstm['Score'], groups=dstm['Teaching_Method'],
                                 alpha=0.05)

#Tukey Visualization

print(tukey_result.plot_simultaneous())

print(tukey_result)
```

```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
import openpyxl
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan

##Part 2

#p2 Import Data#
ole = pd.read_excel("Online Learning and Engagement.xlsx", engine = 'openpyxl')

#p2 Question 6 Exploratory Data Analysis.
print(ole.columns.tolist())
print(ole.describe(include="all").T)

ole_stats = ole.groupby("Platform")["Engagement"].agg(['mean','std'])

print(ole_stats)

#Build Boxplot
sns.set_style("whitegrid")
sns.boxplot(
    data=ole,
    x='Platform',
    y='Engagement'
)
plt.title("Engagement of Students from Different learning platforms\nStudent Engagement Test",
          loc='center',
          fontsize=16)
plt.show()

#p2 Question 7 ANOVA
ole_mod = ols('Engagement ~ Platform+Day+Section', data=ole).fit()
print(ole_mod.summary())
#Residuals
residuals = ole_mod.resid

#QQ plot
sm.qqplot(residuals, line='s')
plt.title("QQ Plot of Residuals")
plt.show()

#S-W test
shapiro = stats.shapiro(residuals)
print(shapiro)

#Test stat and p-value
sw_stat = shapiro.statistic
sw_pval = shapiro.pvalue
print(f"Shapiro-Wilk Test Statistic: {sw_stat:.4f}, p-value: {sw_pval:.4f}")
if sw_pval > 0.05:
    print("✅ Residuals appear to be normally distributed (fail to reject H0).")
else:
    print("❌ Residuals do not appear to be normally distributed (reject H0).")

#p2 Question 8 Homogeneity of Variance

standardized_residuals = ole_mod.get_influence().resid_studentized_internal

#Get the fitted values

fitted_values = ole_mod.fittedvalues

#Create the plot

plt.figure(figsize=(8, 6))
sns.scatterplot(x=fitted_values, y=standardized_residuals)
plt.axhline(y=0, linestyle='dashed', color='red')
plt.axhline(y=3, color='blue')
plt.axhline(y=-3, color='blue')
plt.xlabel("Predicted y's")
plt.ylabel("Standardized Residuals")
plt.title("Standardized Residuals vs. Fitted Values")
plt.style.use('classic')
plt.show()

#B-P Test

bp_test = het_breuschpagan(ole_mod.resid,ole_mod.model.exog)

#Extract F-statistic and p-value

f_statistic = bp_test[2]
f_p_value = bp_test[3]

print(f"F-statistic: {f_statistic}")
print(f"F p-value: {f_p_value}")
if f_p_value > 0.05:
    print("✅ Residuals appear to have constant variance (fail to reject H0).")
else:
    print("❌ Residuals do not appear to have constant variance (reject H0).")

#p2 Question 9

## Print ANOVA Table ##
ole_tab = sm.stats.anova_lm(ole_mod,typ=3)
## Add Significance Column ##
ole_tab['Sig'] = np.where(ole_tab['PR(>F)']<0.05,'Yes','No')
## Print only the main effect and blocking factor rows ##
print(ole_tab.loc[['Platform','Section','Day']])

#p2 Question 10

tukey_result = pairwise_tukeyhsd(endog=ole['Engagement'], groups=ole['Platform'], alpha=0.05)

tukey_result.plot_simultaneous()

print(tukey_result)
#
```
