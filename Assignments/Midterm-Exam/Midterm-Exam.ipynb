**Question 1.** Give a reason why one-factor CRD wouldn't generally be appropriate for studies involving human participants.

One factor CRD's assume that the experimental units are homogenous. This is not the case in human participants due to inherent diversity. Human participants carry many individual differences and lurking variables that could significantly influence the response variable.

**Question 2.** What is the purpose of post-hoc tests?

The purpose of post-hoc tests are to determine precisely which group differs after ANOVA has found statistically significant results. This is because ANOVA does not specify in its results. Pairwise comparisons are usually run in post-hoc tests, which help differentiate significant groups and reduce the probability of making a Type 1 error.

**Question 3.** Explain the difference between RBD and LSD.

Randomized block design controls for one known source of variability, while Latin Square Design controls for two sources of variability but they have to be one blocking factor in each row and each column.

**Question 4.** Explain why randomization is important in designing experiments.

Randomization is important in designing experiments to eliminate bias, conscious or unconscious. It also helps ensure that unknown or lurking variables are distributed "equally" across the treatment groups.

**Question 5.** Explain why replication is important in designing experiments.

Replication is important in designing experiments as it leads to an increase in precision of the estimate of the treatment, decreasing the standard error of the mean. It also ensures that the results were not just a "fluke" and ensures the observed effects are consistent and not just a product of chance.


```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
import openpyxl
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan


#p1 Import Data#
ctrl = pd.read_excel("Video Game Controllers.xlsx", engine = 'openpyxl')

#p1 Question 6 Exploratory Data Analysis.
print("Question 6: Exploratory Data Analysis")
print(ctrl.columns.tolist())
print(ctrl.describe(include="all").T)

print(ctrl.info())

ctrl_stats = ctrl.groupby("Material")["Comfort"].agg(['mean','std'])

print(ctrl_stats)

mtrl = ctrl['Material'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(mtrl.index, mtrl.values, color='green')
plt.title('Question 6 Data Integrity: Materials')
plt.xlabel('Material')
plt.ylabel('Count')
plt.show()

cmft = ctrl['Comfort'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(cmft.index, cmft.values, color='deeppink')
plt.title('Question 6 Data Integrity: Comfort')
plt.xlabel('Comfort')
plt.ylabel('Count')
plt.show()

#Box plot of Comfort by Material
plt.figure(figsize=(10, 6))
sns.boxplot(x='Material', y='Comfort', data=ctrl)
plt.title('Question 6 Additional Visualization: Comfort Scores by Material Type')
plt.xlabel('Material Type')
plt.ylabel('Comfort Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

#Question 7 One-way ANOVA#
f_statistic, p_value = stats.f_oneway(ctrl['Comfort'][ctrl['Material'] == 'Metal'],
                                    ctrl['Comfort'][ctrl['Material'] == 'Plastic'],
                                    ctrl['Comfort'][ctrl['Material'] == 'Silicone'],
                                    ctrl['Comfort'][ctrl['Material'] == 'Rubber'])
#Question 9#
print("Question 9: One-way ANOVA")
print(f"F-statistic: {f_statistic}, P-value: {p_value}")

#Question 7 Normality Tests#
# Visual method: Q-Q plots for each group
for group_name, group_data in ctrl.groupby('Material'):
    plt.figure(figsize=(6, 4))
    sm.qqplot(group_data['Comfort'], line='s', ax=plt.gca())
    plt.title(f'Question 7 Normality Tests: Q-Q Plot for Controller (Material: {group_name})')
    plt.show()

# Testing method: Shapiro-Wilk test for each group
for group_name, group_data in ctrl.groupby('Material'):
    shapiro_statistic, shapiro_p_value = stats.shapiro(group_data['Comfort'])
    print("Question 7 Normality Test: Shapiro-Wilk test")
    print(f"Shapiro-Wilk Test for Comfort (Material : {group_name} ): Statistic={shapiro_statistic:.3f}, P-value={shapiro_p_value:.3f}")

#Question 8#
#visual method: Boxplot to inspect variance across groups
plt.figure(figsize=(8, 6))
sns.boxplot(x='Material', y='Comfort', data=ctrl)
plt.title('Question 8 Homogeneity of Variance: Boxplot')
plt.xlabel('Material')
plt.ylabel('Comfort')
plt.show()

#Testing method: Levene's test
#Extract IBU values for each material group
metal = ctrl['Comfort'][ctrl['Material'] == 'Metal']
plastic = ctrl['Comfort'][ctrl['Material'] == 'Plastic']
rubber = ctrl['Comfort'][ctrl['Material'] == 'Rubber']
silicone = ctrl['Comfort'][ctrl['Material'] == 'Silicone']

Levene_statistic, Levene_p_value = stats.levene(metal, plastic, rubber, silicone)
print("Question 8 Homogeneity of Variance: Levene's Test")
print(f"Levene's Test Statistic: {Levene_statistic}, P-value: {Levene_p_value}")

#Question 10 Tukey's HSD post-hoc test#
#COmvine the comfort and material lavels for the pairwise_hsd function
tukey_results = pairwise_tukeyhsd(endog=ctrl['Comfort'], groups=ctrl['Material'], alpha=0.05)
print("Question 10: Tukey's HSD post-hoc test")
print(tukey_results)

#Tukey Visualization

tukey_results.plot_simultaneous()
plt.show()

print(tukey_results)
```

```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
import openpyxl
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan


#p2 Import Data#
nvrs = pd.read_excel("Nursing VR Study.xlsx", engine = 'openpyxl')

#p2 Question 6 Exploratory Data Analysis.
print("Question 6: Exploratory Data Analysis")
print(nvrs.columns.tolist())
print(nvrs.describe(include="all").T)

print(nvrs.info())

nvrs_stats = nvrs.groupby("Experience")["Score"].agg(['mean','std'])

print(nvrs_stats)

#plot
sns.set_style("whitegrid")

sns.boxplot(
    data=nvrs,
    x='Experience',
    y='Score',
    hue='Method',
    palette='pastel'
)
plt.title('Question 6 Data Integrity: Experience',
          loc='center',
          fontsize=16)
plt.show()

#p2 Question 7 ANOVA
print("Question 7")
#fit model
nvrs_mod = ols('Score ~ Experience + Method', data= nvrs).fit()
print(nvrs_mod.summary())

nvrs_table=sm.stats.anova_lm(nvrs_mod, type = 3)
print(nvrs_table)

#check residuals
nvrs_residuals = nvrs_mod.resid

#QQ plot
sm.qqplot(nvrs_residuals, line='s')
plt.title('Question 8 Normality Tests: Q-Q Plot')
plt.show()

#Shapiro Wilk test
shapiro_test = stats.shapiro(nvrs_residuals)
print(shapiro_test)

#Test stat
shapiro_stat = shapiro_test.statistic

#p-value
shapiro_p = shapiro_test.pvalue

print(f"Shapiro-Wilk Test Statistic: {shapiro_stat:.4f}")
print(f"Shapiro-Wilk Test P-Value: {shapiro_p:.4f}")
if shapiro_p < 0.05:
    print("The residuals are not normally distributed (reject H0).")
else:
    print("The residuals are normally distributed (fail to reject H0).")

#p2 Question 8 Homogeneity of Variance
print("Question 8: Homogeneity of Variance")
#Standardize residuals

standardized_residuals = nvrs_mod.get_influence().resid_studentized_internal

fitted_values = nvrs_mod.fittedvalues

#Create plot

plt.figure(figsize=(8,6))
sns.scatterplot(x=fitted_values, y=standardized_residuals)
plt.axhline(0, color='red', linestyle='--')
plt.axhline(3, color='blue', linestyle='--')
plt.axhline(-3, color='blue', linestyle='--')
plt.title("Standardized Residuals vs. Fitted Values", fontsize=16)
plt.xlabel("Fitted Values")
plt.ylabel("Standardized Residuals")
plt.show()

#B-P test

bp_test = het_breuschpagan(nvrs_mod.resid, nvrs_mod.model.exog)

#F-stat

bp_stat = bp_test[0]

#p-value

bp_p = bp_test[1]

print(f"F-statistic: {bp_stat}")
print(f"F p-value: {bp_p}")
if bp_p < 0.05:
    print("Heteroscedasticity detected (reject H0). Constant Variance not assumed.")
else:
    print("No heteroscedasticity detected (fail to reject H0). Constant Variance assumed.")

#Question 9#
print("Question 9: Anova table")
print(nvrs_table.loc[["Experience", "Method"]])

#p2 Question 10 Tukey HSD
print("Question 10: Tukey's HSD")
tukey_result = pairwise_tukeyhsd(endog=nvrs['Score'], groups=nvrs['Method'],
                                 alpha=0.05)

#Tukey Visualization

tukey_result.plot_simultaneous()
plt.show()

print(tukey_result)
```
